### Training a TransE embeddings model

#### Steps
- Load the reference dataset (filtered in the previous stage)
- Split the data in to train/test/validation
- Create TransE model pipline
- Train and save.


```python
# !pip install pykeen rdflib torch --quiet
```

```python
import torch
import numpy as np
from rdflib import Graph, URIRef, RDF
from pykeen.triples import TriplesFactory
from pykeen.pipeline import pipeline
```

```python
REFERENCE_FILE = "/content/dbpedia-reference-kg.nt"
reference_knowledge_graph = Graph()
reference_knowledge_graph.parse(REFERENCE_FILE)
```

```python
reference_data_numpy = np.array(list(reference_knowledge_graph), dtype=str)
reference_data_pykeen = TriplesFactory.from_labeled_triples(reference_data_numpy)
reference_data_pykeen.create_inverse_triples = True
```

```python
reference_data_pykeen
```

```python
training, testing, validation = reference_data_pykeen.split([0.8, 0.1, 0.1])
# training, testing, validation = reference_data_pykeen.split([0.85, 0.075, 0.075])

result = pipeline(
    training=training,
    testing=testing,
    validation=validation,
    model='TransE',
    model_kwargs={
        'embedding_dim': 50
    },
    epochs=5
)

result.save_to_directory("trans-e-embeddings")
model = result.model
```
